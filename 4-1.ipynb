### 데이터 준비하기

```python
import pandas as pd

fish = pd.read_csv('https://bit.ly/fish_csv_data')
fish.head()
```

```python
print(pd.unique(fish['Species']))
```

```python
fish_input = fish[['Weight','Length','Diagonal','Height','Width']].to_numpy()
```

```python
print(fish_input[:5])
```

```python
fish_target = fish['Species'].to_numpy()
```

```python
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(
    fish_input, fish_target, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
ss.fit(train_input)
train_scaled = ss.transform(train_input)
test_scaled = ss.transform(test_input)
```

### k-최근접 이웃 분류기의 확률 예측

```python
from sklearn.neighbors import KNeighborsClassifier

kn = KNeighborsClassifier(n_neighbors=3)
kn.fit(train_scaled, train_target)

print(kn.score(train_scaled, train_target))
print(kn.score(test_scaled, test_target))
```

```python
print(kn.classes_)
```

```python
print(kn.predict(test_scaled[:5]))
```

```python
import numpy as np

proba = kn.predict_proba(test_scaled[:5])
print(np.round(proba, decimals=4))
```

```python
#5번째 샘플의 최근접 이웃
distances, indexes = kn.kneighbors(test_scaled[4:5])
print(train_target[indexes])
```

```python
distances, indexes = kn.kneighbors(test_scaled[[4]])
print(train_target[indexes])
```

```python
from sklearn.neighbors import KNeighborsClassifier

kn = KNeighborsClassifier(n_neighbors=4)
kn.fit(train_scaled, train_target)

print(kn.score(train_scaled, train_target))
print(kn.score(test_scaled, test_target))
```

```python
kn = KNeighborsClassifier()
kn.fit(train_scaled, train_target)

for n in range(4, 40):
    # 최근접 이웃 개수 설정
    kn.n_neighbors = n
    # 점수 계산
    score = kn.score(train_scaled, train_target)
    # 100% 정확도에 미치지 못하는 이웃 개수 출력
    if score < 1:
        print(n, score)
        break
```

### 로지스틱 회귀로 이진 분류 수행하기

```python
char_arr = np.array(['A', 'B', 'C', 'D', 'E'])
print(char_arr[[True, False, True, False, False]])
```

```python
bream_smelt_indexes = (train_target == 'Bream') | (train_target == 'Smelt')
train_bream_smelt = train_scaled[bream_smelt_indexes]
target_bream_smelt = train_target[bream_smelt_indexes]
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(train_bream_smelt, target_bream_smelt)
```

```python
print(lr.predict(train_bream_smelt[:5]))
```

```python
print(lr.predict_proba(train_bream_smelt[:5]))
```

## 음성 양성 클래스 구분

```python
print(lr.classes_)
```

### 방정식에 사용되는 기울기(가중치,계수) 및 절편

```python
print(lr.coef_, lr.intercept_)
```

### 로지스틱 회귀 모델로 양성 클래스에 대한 z계산

```python
decisions = lr.decision_function(train_bream_smelt[:5])
print(decisions)
```

### 양성 클래스에 대한 확률 계산

```python
from scipy.special import expit

print(expit(decisions))
```

### 로지스틱 회귀로 다중 분류 수행하기

훈련
반복적인 알고리즘을 사용함
기본값은 max_iter=100 으로서 불충분하기 때문에 1000으로 상향조정함
릿지 회귀 방식(L2 규제)와 같이 계수의 제곱을 규제함
그러나 매개변수는C이고 작을수록 규제가 커짐
선형회귀에서는 alpha이고 클수록 규제 강도가 세짐
기본값은 C=1이지만 규제를 느슨하게 하기 위해 C=20으로 상향조정함
점수 확인
예측
확률(소수점 4자리에서 반올림)
클래스
계수와 절편

```python
lr = LogisticRegression(C=20, max_iter=1000)
lr.fit(train_scaled, train_target)

print(lr.score(train_scaled, train_target))
print(lr.score(test_scaled, test_target))
```

```python
print(lr.predict(test_scaled[:5]))
```

```python
proba = lr.predict_proba(test_scaled[:5])
print(np.round(proba, decimals=3)) #소수점4자리에서 반올림
```

```python
print(lr.classes_)
```

```python
print(lr.coef_.shape, lr.intercept_.shape)
```

선형방정식에 사용되는 계수와 절편의 형태 확인

5개의 특성을 사용하므로 열은 5개, 타깃이7개이므로 행은 7개, 즉 z를 7번 계산함

```python
decision = lr.decision_function(test_scaled[:5])
print(np.round(decision, decimals=2))
```

axis=1, 각 행, 즉 각 샘플에 대해 소프트맥스를 계산

만약 axis를 지정하지 않는다면 모든 값에 대해 소프트맥스를 계산

```python
from scipy.special import softmax

proba = softmax(decision, axis=1)
print(np.round(proba, decimals=3))
```
